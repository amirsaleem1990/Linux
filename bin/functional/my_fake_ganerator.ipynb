```python
# duplicated rows
x_3 = x.append(x).reset_index(drop=True)
rows_series = x_3[[0,1]].duplicated()
rows = rows_series[rows_series].index.values
x_3.style.apply(lambda x: ['background: green' if x.name in rows else '' for i in x], axis=1)
```

```python
def highlight_vals(row, cols=[0, 3], color1='green', color2="blue"):
    a, b = cols
    styles = {col: '' for col in row.index}
    if row[a] > 0.25 * row[b]:
        styles[a] = 'background-color: %s' % color1
        styles[b] = 'background-color: %s' % color2
    return styles

x.style.apply(lambda x: highlight_vals(x), axis=1)
```
